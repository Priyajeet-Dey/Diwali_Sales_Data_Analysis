```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.columns
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
ax = sns.countplot(x = 'Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', hue='Gender', data = sales_gen)
```

```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', hue='Age Group', data = sales_age)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State', y = 'Orders', hue='State', data = sales_state)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State', y = 'Amount', hue='State', data = sales_state)
```

```python
ax = sns.countplot(x = 'Marital_Status', hue= 'Marital_Status',  data=df)
sns.set(rc={'figure.figsize':(6,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', hue='Gender', data = sales_state)
```

```python
ax = sns.countplot(x = 'Occupation', hue= 'Occupation',  data=df)
sns.set(rc={'figure.figsize':(5,20)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Occupation', y = 'Amount', hue='Occupation', data = sales_state)
```

```python
ax = sns.countplot(x = 'Product_Category', hue= 'Product_Category',  data=df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_Category', y = 'Amount', hue='Product_Category', data = sales_state)
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_ID', y = 'Orders', hue='Product_ID', data = sales_state)
```

Conclusions: Married women aged 26-35 from Uttar Pradesh, Maharashtra, and Karnataka who work in the IT, healthcare, and aviation sectors show a higher propensity for purchasing products from food, clothing, and electronics categories. This demographic likely has disposable income and specific lifestyle needs that drive their purchasing decisions in these areas. They are attracted to food products for convenience and health benefits, clothing for professional and personal attire, and electronics for both work-related and personal use. Understanding the buying habits of this group can help businesses tailor their marketing strategies to effectively target and meet their needs.
